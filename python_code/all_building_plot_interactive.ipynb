```python
import pandas as pd

import geopandas as gpd
df = pd.read_csv('/Users/ryanng/Desktop/nyc_pluto_22v3_1_csv/pluto_22v3_1.csv')

df_selected = df[['lotarea', 'bldgarea','zonedist1','latitude','longitude','cd']]
```

```python
df_selected.dropna(subset=['bldgarea'], inplace=True)
df_selected
```

```python
df_selected.dropna(subset=['zonedist1'], inplace=True)
df_selected
```

```python
def get_first_letter(string):
    return string[0]
df_selected['column_name'] = df_selected['zonedist1'].apply(get_first_letter)
```

```python
gdf = gpd.GeoDataFrame(df_selected, geometry=gpd.points_from_xy(df_selected.longitude, df_selected.latitude))
gdf
```

```python
gdf['density']=gdf['bldgarea']/gdf['lotarea']
gdf
```

```python
# 创建一个字典，用于将区域代码转换为区域名称

df = gdf.dropna(subset=['cd'])
df['cd'] = df['cd'].astype(int)

boro_codes = {
    1: 'MN',
    2: 'BX',
    3: 'BK',
    4: 'QN',
    5: 'SI'
}

# 根据'BoroCD'列的值，将dataframe分为子dataframe，并画出图表
for code in df['cd'].unique():
    boro = boro_codes[int(str(code)[0])]
    cd_num = code % 100  # 获取后两位数字
    sub_df = df[df['cd'] == code]  # 获取子dataframe
    
    # 绘制图表
    fig, ax = plt.subplots(figsize=(10, 10))
    
    plt.title(f'{boro}{cd_num}')  # 设置图表标题
    filename = f'{boro}{cd_num}.jpg'
    sub_df.plot(ax=ax, column='density', cmap='Blues', legend=True).get_figure().savefig(filename)
    #mplleaflet.save_html(fig, fileobj=filename)
    #plt.savefig(f'{boro}{cd_num}.png')  # 保存图像
    plt.clf()  # 清除图像
```

```python
import folium
import numpy as np
from folium.plugins import HeatMap
```

```python
lat = np.array(gdf["latitude"][0:len(gdf)])
lon = np.array(gdf["longitude"][0:len(gdf)])
density = np.array(gdf["density"][0:len(gdf)],dtype=float)
data1 = [[lat[i],lon[i],density[i]] for i in range(len(gdf))]
```

```python
a=gdf.latitude.mean()
b=gdf.longitude.mean()
```

```python
categories = gdf['column_name'].unique()
```

```python
map_osm = folium.Map(location=[a,b],zoom_start=10)
HeatMap(data1).add_to(map_osm) 

map_osm.save('park_map2.html')
```

```python
df=gdf[['latitude','longitude','density','column_name']]
df
```

```python
# create a Folium map centered on the mean of the latitudes and longitudes
m = folium.Map(location=[df['latitude'].mean(), df['longitude'].mean()], zoom_start=10)

# create a dictionary to store the HeatMap layers for each category
hm_dict = {}
for category in df['column_name'].unique():
    # create a HeatMap layer for the current category
    data = df[df['column_name'] == category][['latitude', 'longitude', 'density']].values.tolist()
    hm = HeatMap(data=data, name=category)
    # add the HeatMap layer to the dictionary
    hm_dict[category] = hm
    # add the HeatMap layer to the map
    hm.add_to(m)

# create a LayerControl object to allow the user to select the category as the legend
folium.LayerControl(collapsed=False, overlay=True, control=False, autoZIndex=True, 
                    name="Legend", position="topright").add_to(m)
for category in df['column_name'].unique():
    # add the HeatMap layer to the LayerControl object
    hm_dict[category].add_to(m)
m.save('park_map3.html')
```
